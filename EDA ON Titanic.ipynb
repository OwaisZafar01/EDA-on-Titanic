## **üìä Univariate Analysis**

Univariate analysis focuses on **understanding the distribution of a single variable**.  
It helps to identify:  
- How the data is **spread**  
- Presence of **outliers**  
- **Most frequent values**

---

### **1Ô∏è‚É£ Numerical Columns**
- Visualized using **Histogram**, **KDE Plot**, and **Box Plot**  
- Example columns: `Age`, `Fare`

**Key insights to observe:**  
1. Mean, Median, and Standard Deviation  
2. Skewness (Left or Right)  
3. Presence of outliers  

---

### **2Ô∏è‚É£ Categorical Columns**
- Visualized using **Countplot**, **Bar Chart**, and **Pie Chart**  
- Example columns: `Sex`, `Pclass`, `Embarked`

**Key insights to observe:**  
1. Most frequent categories  
2. Rare categories  
3. Class imbalance (if any)  

---

### **3Ô∏è‚É£ Mixed / String Columns**
- Columns like `Name`, `Ticket`, `Cabin`  
- Mostly analyzed through **counts or unique values**  
- Detailed statistical visualization is usually not meaningful  

---

## **üìä Bivariate Analysis**

Bivariate analysis means studying the **relationship between two columns** in a dataset.  
It helps us understand:  
- How one column affects or relates to another  
- Patterns or trends between the two columns  
- Any unusual points or imbalances  

---

### **1Ô∏è‚É£ Numerical ‚Üî Numerical**
- Both columns are **numbers** 
- Use **Scatter Plot**, **2D Histogram**, or **2D KDE Plot** to see the relationship  
- Can also check **correlation** to see if they increase or decrease together  

**Things to look for:**  
1. Are the numbers **moving together** (positive or negative relationship)?  
2. Any **clusters or patterns**?  
3. Any **outliers** that don‚Äôt follow the trend?  

---

### **2Ô∏è‚É£ Numerical ‚Üî Categorical**
- One column is **number** and the other is **category**   
- Use **Box Plot**, **Violin Plot**, **Bar Plot** to compare  

**Things to look for:**  
1. Which category has **higher or lower numbers**  
2. **Spread of data** in each category  
3. If there are **outliers** in some categories  

---

### **3Ô∏è‚É£ Categorical ‚Üî Categorical**
- Both columns are **categories**  
- Use **Count Plot**, **Stacked Bar Plot**, or **Heatmap**  

**Things to look for:**  
1. Which combination appears **most often**  
2. Any **imbalance** in the counts  
3. Very **rare combinations** of categories  

---

**Next Step:**  
- Pick any two columns, make a chart, and see **how they are related**.  
- Write down **your observations** clearly.

```python
df['Family_size'] = df['SibSp'] + df['Parch'] + 1
df.head()
```

```python
# family type

def transform_family_size(num):
    if num == 1:
        return 'Alone'

    elif num > 1 and num < 5:
        return 'Small'

    else:
        return 'Large'
```

```python
df['Family_Type'] = df['Family_size'].apply(transform_family_size)
df.head()
```

```python
crosstab = pd.crosstab(df['Survived'], df['Family_Type'], normalize='columns')*100
crosstab 
```

```python
plt.figure(figsize=(7,5))
sns.heatmap(crosstab, 
            annot=True,             
            fmt='.1f',              
            cmap='viridis',       
            cbar_kws={'label': 'Percentage %'}, 
            linewidths=0.5, 
            linecolor='white')

plt.title('Survival Rate by Family Type', fontsize=16, fontweight='bold', color='#333333')
plt.xlabel('Family Type', fontsize=12)
plt.ylabel('Survived (0 = Died, 1 = Survived)', fontsize=12)
plt.xticks(rotation=0)
plt.show()
```


### **Observation ‚Äì Survival Rate by Family Type**

- *Passengers traveling with **small families** had the **highest survival chance (~57.9%)***.  
- *Passengers traveling **alone** had a **much lower survival rate (~30.3%)***.  
- *Passengers with **large families** were the **riskiest (~16.1% survival rate)***.  
- **Overall Trend:**  
  *Survival chances decrease as family size increases ‚Üí*  
  **Small > Alone > Large**
